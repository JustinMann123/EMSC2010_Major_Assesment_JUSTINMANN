# ***Determining the Empirical Elastic Admittance of the Greenland Crust.***

The Greenland Ice Sheet is the second largest body of ice on Earth, holding enough water to raise global sea levels by approximately seven meters if fully melted. While complete collapse is not of immediate concern, Greenland is already one of the largest contributors to current-day sea level rise, and the rate of mass loss has accelerated over recent decades. Understanding exactly how much ice Greenland is losing, and how quickly, is therefore one of the more consequential problems in modern climate science.


Along with ice melt, there are multiple other processes that occur and hold the key to answering these consequential problems. Of the many, ongoing research into how the Earth, and more precisely the Greenland lithosphere, is responding in modern times is of particular interest. The two main processes that are of interest to researchers are Glacial Isostatic Adjustment (GIA) and elastic deformation.


GIA is the slow, ongoing rebound of the Earth's crust in response to ice lost since the Last Glacial Maximum. Elastic deformation, on the other hand, is the immediate crustal response to modern ice mass change, producing predictable vertical oscillations as ice surface elevation changes. Accurately partitioning these signals is essential for using bedrock displacement as a mass balance constraint.


This vertical displacement is measurable through GPS networks, offering a method to quantify both GIA and elastic deformation. However, isolating the ice-driven signal requires separating the two distinct mechanisms of vertical bedrock motion. This is an issue as GIA and elastic deformation are affected by multicollinearity. As a result, to isolate one or the other, modeled GIA must be removed from the signal.



**Why is this important?**

Currently, ice mass balance (the net difference between the ice gains (via snowfall) and the ice it loses (through melting, runoff, and calving into the ocean)) studies are measured using satellite gravity data from GRACE. GRACE detects mass redistribution through changes in Earth's gravity field. Both GIA and elastic deformation must be considered in mass balance equations. Any vertical deformation will influence the gravitational field that GRACE senses. (A drop in bedrock/mantle due to either GIA or elastic deformation, or both, will lessen the gravitational field's strength, and vice versa when bedrock/mantle moves upwards). A decrease in gravitational forces, if caused by vertical displacement (and no ice was lost), would be seen as mass lost, as GRACE cannot distinguish the source of the decreased gravity. Even more consequentially, if we have ice mass loss, and as a result there is an increased uplift of bedrock, the gravity fields would make it appear that there is little mass change, even though mass has been lost in the form of ice melt (which is what we care about for sea level rise studies).


As a result, the more accurately both the GIA and elastic deformation signals can be removed from GRACE or any mass balance methodology (there are many different methods, such as using altimetry, which would require knowledge about how much the ice is moving from ice loss versus bedrock movement), the closer the estimate to true mass balance.

**This raises a critical question:** Can we isolate one of the two signals (elastic deformation) and provide critical information on the elastic admittance that can be used to estimate how much vertical movement occurs for a specific change in mass? And can this admittance be used to determine the mass balance of Greenland?

**Secondary question:** Is seasonal elastic deformation admittance different from long-term climate-driven admittance?

We hypothesize that long-term elastic deformation occurs at a different rate than seasonal, due to different rheological processes.

As mentioned previously, GIA is one of two mechanisms that contribute to the vertical displacement of the Earth's crust. However, removing GIA from the vertical signal is not straightforward. GNET GPS only shows one elevation point in time, thus you cannot effectively attribute either GIA or elastic deformation, or isolate one from the other. This poses a massive issue as, to effectively isolate both the seasonal wiggle of crustal deformation and the long-term uplift caused by climate-driven ice melt, GIA needs to be removed.

To do this, I have incorporated a GIA model that provides a dataset determining the rate of displacement from GIA at each GNET station. Removing this rate from each observation effectively leaves us with the somewhat linear signal, with a seasonal wiggle, that we expect with elastic deformation.


# **To perform the removal of GIA from the GNET signal, a few steps must be undertaken prior to spatially averaging.**


**Methods**

1.  **Spatially match GIA rates to GNET stations**

    - A nearest-neighbour spatial search (`cKDTree`) was used to pair each GNET station with its closest VLM (Vertical Land Motion) site, assigning a station-specific GIA rate.

    - This was required as the GNET data utilized site names, while the VLM data utilized site IDs (e.g., 1, 2, 3...).


2.  **Temporal Aggregation (GNET):**

    - GNET observations, initially daily measurements, were resampled to monthly means to align with GRACE's monthly measurements.

    - Raw daily GNET vertical displacement (up) was sliced to the study period and resampled to monthly averages to align with GRACE temporal resolution.


3.  **Normalize each station**

    - Each station's monthly time series was mean-subtracted to remove elevation offsets that occur depending on when the GPS sensor was placed. Thus, all stations represent relative change from their own average.

4.  **Subtract the GIA signal**

    - For each station, a linear GIA rate was derived from the GIA model, which for a given GNET station would have an 'x' mm/yr of displacement. Removing this rate was done by dividing by 12 months to give the displacement for that month.


5.  **Spatial Averaging (GNET):**

    - The normalized GNET signal, consisting of individual station data, was averaged across all stations for each month to represent Greenland as a single 1-dimensional measurement.

```python
from scipy.spatial import cKDTree

# --- Load VLM ---

url = 'https://raw.githubusercontent.com/JustinMann123/EMSC2010_Major_Assesment_JUSTINMANN/main/VLM_sites.dat'
vlm = pd.read_csv(url, sep=r'\s+', skiprows=1,
                   names=['Site', 'Lon', 'Lat', 'VLM_3D', 'Sig_3D', 'VLM_1D', 'Sig_1D'])

# --- Match each GNET station to nearest VLM station by lat/lon ---
gnet_coords = np.column_stack([gnet['lon'].values, gnet['lat'].values])
vlm_coords  = np.column_stack([vlm['Lon'].values, vlm['Lat'].values])

tree = cKDTree(vlm_coords)
distances, indices = tree.query(gnet_coords)

station_names = gnet['station'].values
gia_lookup    = pd.Series(vlm['VLM_1D'].values[indices], index=station_names)
```

```python
# --- Build two raw unnormalised arrays ---
study_start = '2002-01-01'
study_end   = '2025-01-01'

time       = pd.to_datetime(gnet['time'].values)
up         = gnet['up'].values
time_index = pd.date_range(study_start, study_end, freq='MS')

raw_array         = []   # unnormalised, no GIA removed
gia_removed_array = []   # unnormalised, GIA removed

for i, stn in enumerate(station_names):

    # slice to study period and resample to monthly
    series = pd.Series(up[i], index=time)
    series = series[(series.index >= study_start) & (series.index < study_end)]
    monthly = series.resample('MS').mean().reindex(time_index)

    # store raw
    raw_array.append(monthly.values)

    # build GIA signal and subtract
    rate      = gia_lookup[stn]                               # mm/yr
    t_years   = (time_index - time_index[0]).days / 365.25    # elapsed years
    gia_signal = rate * t_years                               # mm

    gia_removed_array.append((monthly - gia_signal).values)

raw_array         = np.array(raw_array)          # shape (70, n_months)
gia_removed_array = np.array(gia_removed_array)  # shape (70, n_months)

print(f"raw_array shape:         {raw_array.shape}")
print(f"gia_removed_array shape: {gia_removed_array.shape}")
```

```python
# --- Plot both for a visual check ---
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

for stn_data in raw_array:
    axes[0].scatter(time_index, stn_data, s=2, alpha=0.3, color='steelblue')
axes[0].set_ylabel('Up (mm)')
axes[0].set_title('Raw unnormalised GNET (all stations)')
axes[0].grid(True)

for stn_data in gia_removed_array:
    axes[1].scatter(time_index, stn_data, s=2, alpha=0.3, color='steelblue')
axes[1].set_ylabel('Up (mm)')
axes[1].set_title('GIA removed unnormalised GNET (all stations)')
axes[1].grid(True)

axes[1].set_xlabel('Time')
plt.tight_layout()
plt.show()
```

# **Why do we need to normalise?**

Each of the 61 GNET stations has a different absolute elevation one station might sit at 150mm and another at 200mm above its reference point simply due to where it was installed. Normalising by subtracting each station's own mean removes these arbitrary offsets so all stations are expressed as change from their individual average rather than absolute elevation. This allows the ensemble mean to represent a Greenland-wide signal of relative elevation change, rather than being dominated by whichever stations happen to sit at the highest absolute elevation.



```python
raw_array         = []
gia_removed_array = []

for i, stn in enumerate(station_names):

    # 1. slice and resample
    series = pd.Series(up[i], index=time)
    series = series[(series.index >= study_start) & (series.index < study_end)]
    monthly = series.resample('MS').mean().reindex(time_index)

    # 2. drop stations with less than 60 months
    if monthly.dropna().__len__() < 60:
        continue

    # 3. normalise FIRST (remove mean offset)
    normalised = monthly - monthly.mean()
    raw_array.append(normalised.values)

    # 4. THEN remove GIA from the normalised signal
    rate      = gia_lookup[stn]
    t_years   = (time_index - time_index[0]).days / 365.25
    gia_signal = rate * t_years

    gia_removed_array.append((normalised - gia_signal).values)

raw_array         = np.array(raw_array)
gia_removed_array = np.array(gia_removed_array)
```

##**GRACE data transformation Methods**


The second piece of the puzzle needed to determine elastic admittance is ice mass change, which is where GRACE comes in. Before jumping into the analysis there were two things that needed to be sorted out first.

**Unit Conversion**

(kg/m² -> Gt)

The Grace dataset from (https://data1.geo.tu-dresden.de/gis_gmb/) has represented mass change in kg/m², which is conceptually fine but would result in admittance values that are pretty hard to make sense of, ie way to large.

To fix this, the area of each mascon grid cell was pulled from the grid spacing, which then allowed the conversion from kg/m² to total kg (kg/m² × m² = kg).

From there, dividing by 1×10¹² gives you Gigatons (Gt w.e.), which is a much more interpretable unit when talking about Greenland-scale ice mass change.


**Spatial Integration**


GRACE mascon values are anomalies, essentially how much each month has deviated from the long term mean — spread across a grid of individual cells.

To get a single number representing all of Greenland each month, all the mascon values were summed together. This effectively collapses Greenland down to a 1D mass change signal that can be directly compared to the spatially averaged GNET signal in the next steps.


```python
grace_mass = grace_grid['dm'].values
grace_time = grace_grid['time'].values

## ai assited with exctracting the cell size a the for loop


dx = np.abs(np.diff(grace_grid.x.values).mean())
dy = np.abs(np.diff(grace_grid.y.values).mean())
cell_area_m2 = dx * dy

time_series = []
for t in range(grace_mass.shape[0]):
    kgm2 = grace_mass[t, :, :]
    total_kg = np.nansum(kgm2 * cell_area_m2)
    total_Gt = total_kg / 1e12
    time_series.append(total_Gt)


plt.figure(figsize=(8, 6))
plt.scatter(grace_time, time_series, marker='o', linestyle='-', s=4, label='Greenland Total Mass Change')
plt.xlabel('YEAR (GRACE STUDY PERIOD)')
plt.ylabel('Total Mass Change (Gt w.e.)')
plt.title('Greenland Total Monthly Mass Change from GRACE')
plt.grid(True)
plt.legend()
plt.show()
```

# **Primary objective 3:** Separating the observed vertical displacement signal into seasonal elastic deformation and long-term elastic deformation

**Methods and reasoning**

- As mentioned previously, GNET measurement records the total vertical displacement of the bedrock. Because we have removed GIA, we can now estimate the elastic admittance of Greenland. However, we cannot simply perform a multiple regression of the GIA-removed signal and GRACE to estimate a posterior for elastic admittance.

- This is due to how deep the stress (from mass gain or loss) penetrates into the Earth's layered structure. For climate-driven elastic deformation over the 16-year study period, the deeper mantle layers would be responding differently when compared to the seasonal signal. The seasonal signal is most likely too quick to impact the deeper layer; therefore, I suspect it will have a different admittance rate.

- As a result, I need to isolate both the annual signal and the climate-driven signal to separate both deformation signals in the GNET data.


**Methods**

1.  **Determining the best-fitting polynomial & isolating seasonal and long-term trends**

    - A leave-one-out Bayesian model comparison, as used previously, was applied to both the GNET and GRACE time series to confirm which polynomial best fit both datasets.

    - Using the best-fitting polynomial for both GRACE and GNET, the trend is then used to detrend the data. This is done by subtracting the newly formed polynomial trend from the transformed data. The result is detrended GNET and GRACE time series that strictly represent the seasonal variations in vertical displacement and mass changes.

    - The best-fitting polynomial trend for GRACE and GNET also now represents the long-term climate-change-driven vertical displacement and mass loss across Greenland.

# **leave-one-out Bayesian model comparison for GIA removed GNET data set**

```python
gnet_time = alligned_grace_gnet['decimal_year']
gnet_nan_removed = alligned_grace_gnet['GNET_Avg_Elevation']

plt.plot(gnet_time, gnet_nan_removed)
plt.xlabel('Time')
plt.ylabel('GNET Elevation Change)')
plt.minorticks_on()

data_gnet = pd.DataFrame({"y": gnet_nan_removed}) #define the y-variable
data_gnet["x_scaled"] = (gnet_time-np.mean(gnet_time)) / np.std(gnet_time) #scale the x values to have a mean of
data_gnet["x2_scaled"] = data_gnet["x_scaled"] ** 2 #used for quadratic regression
data_gnet["x3_scaled"] = data_gnet["x_scaled"] ** 3 #used for cubic regression
data_gnet["x4_scaled"] = data_gnet["x_scaled"] ** 4 #used for quartic regression

# Fit a straight-line model (first-order polynomial)
model_linear_gnet = bmb.Model("y ~ x_scaled", data_gnet)
idata_linear_gnet = model_linear_gnet.fit(idata_kwargs={"log_likelihood": True})

# Fit a quadratic model (second-order polynomial)
model_quad_gnet = bmb.Model("y ~ x_scaled + x2_scaled", data_gnet)
idata_quad_gnet = model_quad_gnet.fit(idata_kwargs={"log_likelihood": True})

# Fit a cubic model (third-order polynomial)
model_cubic_gnet = bmb.Model("y ~ x_scaled + x2_scaled + x3_scaled", data_gnet)
idata_cubic_gnet = model_cubic_gnet.fit(idata_kwargs={"log_likelihood": True})

# Fit a quartic model (fourth-order polynomial)
model_quart_gnet = bmb.Model("y ~ x_scaled + x2_scaled + x3_scaled + x4_scaled", data_gnet)
idata_quart_gnet = model_quart_gnet.fit(idata_kwargs={"log_likelihood": True})

#Compare models
comparison_gnet = az.compare({
    "linear": idata_linear_gnet,
    "quadratic": idata_quad_gnet,
    "cubic": idata_cubic_gnet,
    "quartic": idata_quart_gnet
})
print(comparison_gnet) #print the result
```

# **Primary objective 4:** Determine the empirical elastic admittance of Greenland's seasonal ice mass variations & Greenland's admittance of climate change driven mass variations.

**Methods and Reasoning**

To be able to determine the empirical admittance for both of the isolated signals, we must quantify how much vertical displacement (mm) occurs per Gigaton of ice mass change.

**Methods**

- A Bayesian linear regression was fitted in Bambi with detrended GRACE as the predictor and detrended GNET as the response. Bambi samples the full posterior distribution of the slope, with the posterior mean representing the seasonal elastic admittance (mm/Gt), while the 95% HDI captures the uncertainty.

- The same Bayesian linear regression was applied in Bambi, this time using the GRACE cubic trend to predict the GNET cubic trend. The posterior mean of the slope is taken as the long-term climate-driven elastic admittance (mm/Gt) with the 95% HDI capturing the uncertainty.


- The posterior mean for both seasonal and long-term admittance was extracted by averaging across all the draws from the slope coefficient, with the 95% HDI calculated as the narrowest interval containing 95% of the posterior probability mass.

# **Bayesian linear regression - detrended GRACE as the predictor and detrended GNET as the response**

```python

# Assuming grace_detrended_cubic and gnet_detrended_cubic are available from previous cells.
# Create DataFrame for the detrended data
data_detrended_cubic_reg = pd.DataFrame({
    "x": grace_detrended_cubic, # Using detrended GRACE as the predictor (x)
    "y": gnet_detrended_cubic # Using detrended GNET as the response (y)
})

# Fit a linear model using bambi
model_cubic_detrended_regression = bmb.Model("y ~ x", data_detrended_cubic_reg)
# Use appropriate number of draws and chains, default to what was used in other cells.
idata_cubic_detrended_regression = model_cubic_detrended_regression.fit(draws=2000, tune=2000, chains=4)

# Check if idata_cubic_detrended_regression is valid after fitting
if idata_cubic_detrended_regression is None or not hasattr(idata_cubic_detrended_regression, 'posterior'):
    raise ValueError("Model fitting failed or idata_cubic_detrended_regression is invalid.")

# Predict across a range of x values (detrended GRACE data range)
x_range_detrended = np.linspace(grace_detrended_cubic.min(), grace_detrended_cubic.max(), 100)

# Manually construct the mean predictions from the posterior coefficients for the regression line
posterior_intercept = idata_cubic_detrended_regression.posterior["Intercept"].values
posterior_x_coeff = idata_cubic_detrended_regression.posterior["x"].values
posterior_sigma = idata_cubic_detrended_regression.posterior["sigma"].values # Get sigma for Posterior Predictive Samples

# Generate predicted y values for each draw and each x_range point (mean of the regression line)
# predictions_draws_mean will have shape (chains, draws, len(x_range_detrended))
predictions_draws_mean = posterior_intercept[..., np.newaxis] + posterior_x_coeff[..., np.newaxis] * x_range_detrended

# Flatten chain and draw dimensions into a single 'draw' dimension
y_mean_draws = predictions_draws_mean.reshape(-1, len(x_range_detrended))

# --- Manual Posterior Predictive Sampling (PPS) ---
# Reshape sigma to be compatible for broadcasting with mu_flat
posterior_sigma_flat = posterior_sigma.reshape(-1, 1)

# Ensure mu_flat (mean prediction for PPS) is flattened (total_draws, x_points)
mu_for_pps = predictions_draws_mean.reshape(-1, len(x_range_detrended))

# Sample from a normal distribution for each draw, using the corresponding posterior sigma
y_pps_draws = np.random.normal(loc=mu_for_pps, scale=posterior_sigma_flat, size=mu_for_pps.shape)

# Compute mean of the regression lines and HDIs
posterior_mean = y_mean_draws.mean(axis=0) # Mean of the regression lines
hdi_mean = az.hdi(y_mean_draws, prob=0.95, axis=0) # HDI for the regression lines
hdi_pps = az.hdi(y_pps_draws, prob=0.95, axis=0) # HDI for the posterior predictive distribution

# Plot both posterior for regression lines and the observations
plt.figure(figsize=(16, 4)) # Create a new figure for this plot
plt.plot(grace_detrended_cubic, gnet_detrended_cubic, 'ok', label="Detrended Data", markersize=1) # Use detrended data for plotting observations
plt.plot(x_range_detrended, posterior_mean, color="C1", label="Posterior mean")
plt.fill_between(
    x_range_detrended,
    hdi_mean[:, 0],
    hdi_mean[:, 1],
    alpha=0.3,
    color="C1",
    label="95% HDI (mean)",
    edgecolor = None
)

plt.fill_between(
    x_range_detrended,
    hdi_pps[:, 0],
    hdi_pps[:, 1],
    alpha=0.2,
    color="C1",
    label="95% HDI (predictive)",
    edgecolor = None
)

plt.legend()
plt.minorticks_on()
plt.xlabel('Detrended GRACE Total Mass Change (Gt w.e.)')
plt.ylabel('Detrended GNET Average Elevation (mm)')
plt.title('Detrended GNET Elevation vs. Detrended GRACE Mass Change and Regression')
plt.show()
```

# **Limitations**

There were many limitations with this analysis.



**Spatial averaging**

Collapsing all GNET stations and GRACE mascons into a single Greenland-wide signal removes any localised signal and variability. This has a massive impact on accuracy. Assuming Greenland moves at the same admittance is incorrect, as it is well documented that different parts are more stable than other parts.


**GNET coastline bias**

Due to GNET only having GPS stations positioned around the coastline of Greenland (see Figure 1) (This is due to it not being possible to place a GPS on bedrock if there is an ice sheet on top), we are unable to effectively represent how the bedrock in the center of Greenland responds to mass change. Therefore, the admittance rates would not be an accurate reflection of Greenland as a whole.


**GIA model dependency**

GIA models are notorious for being unreliable, due to the difficulty in forming said models. GIA models require complex and poorly understood systems. Firstly, GIA models must take into account the past history of ice sheets from the last glacial maximum (very hard to do) and the rheology of the lithosphere, which is also very hard to quantify.

Therefore, I have a heavy dependence on this GIA model, but spatial averaging may have reduced this dependence.



**GRACE data gaps and limited resoluiton**

The inevitable gaps in GRACE data are also a limitation, but realistically there is nothing I could do to resolve this.
